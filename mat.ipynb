```python
Matplotlib is a low-level library of Python which is used for data visualization. It is easy to use 
and emulates MATLAB like graphs and visualization. This library is built on the top of NumPy arrays. 
consist of several plots like line chart, bar chart, histogram, scatter,line graph etc. 
```

```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scatter plot ")
plt.scatter(x,y )
```

```python
With the subplot() function you can draw multiple plots in one figure:
```

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x,y)
plt.subplot(2,2,1)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x,y)
plt.subplot(2,2,2)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x,y)
plt.subplot(2,2,3)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x,y)
plt.subplot(2,2,4)

plt.plot(x,y)                                                  
                                                  
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and 
heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. 
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.xlabel("company")
plt.ylabel("profit")
plt.title('Data')
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having
properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created 
from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. 
Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box1)

plt.boxplot(box2)

```
